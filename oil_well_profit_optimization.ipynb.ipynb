# 1. Importar librer√≠as y cargar los datos

Cargaremos los datos de las tres regiones para analizarlos. Estos archivos contienen informaci√≥n de los pozos como sus caracter√≠sticas (`f0`, `f1`, `f2`) y el volumen de reservas (`product`).

```python
# Librer√≠as principales
import pandas as pd
import numpy as np

# Visualizaci√≥n
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cargar los datos
data_0 = pd.read_csv('/datasets/geo_data_0.csv')
data_1 = pd.read_csv('/datasets/geo_data_1.csv')
data_2 = pd.read_csv('/datasets/geo_data_2.csv')

# Ver primeros registros de la regi√≥n 0 como ejemplo
data_0.head()
```

# 2. Entrenar y evaluar el modelo para la regi√≥n 0

Dividiremos los datos en entrenamiento y validaci√≥n, entrenaremos un modelo de regresi√≥n lineal, haremos predicciones y evaluaremos el rendimiento con RMSE.

```python
# Dividir en caracter√≠sticas (X) y objetivo (y)
features_0 = data_0.drop(['product', 'id'], axis=1)
target_0 = data_0['product']

# Dividir en entrenamiento y validaci√≥n (75% - 25%)
X_train_0, X_valid_0, y_train_0, y_valid_0 = train_test_split(
    features_0, target_0, test_size=0.25, random_state=42
)

# Entrenar el modelo
model_0 = LinearRegression()
model_0.fit(X_train_0, y_train_0)

# Predecir
predictions_0 = model_0.predict(X_valid_0)
# Evaluar el modelo
rmse_0 = mean_squared_error(y_valid_0, predictions_0, squared=False)
mean_pred_0 = predictions_0.mean()

print(f"RMSE regi√≥n 0: {rmse_0:.2f}")
print(f"Volumen promedio predicho: {mean_pred_0:.2f}")
```

# 5. Preparar datos para c√°lculo de ganancias

Definimos las variables clave del negocio y comparamos el umbral m√≠nimo de producci√≥n (111.1 unidades) con el volumen medio predicho en cada regi√≥n.

```python
# Variables del negocio
BUDGET = 100_000_000         # 100 millones
POZOS = 200
COSTO_POR_POZO = BUDGET / POZOS
INGRESO_POR_UNIDAD = 4500    # $ por mil barriles
UMBRAL_MINIMO = COSTO_POR_POZO / INGRESO_POR_UNIDAD

print(f"Cada pozo debe producir al menos {UMBRAL_MINIMO:.1f} mil barriles para no tener p√©rdidas.")

# Comparar con los valores promedio de cada regi√≥n
print(f"Promedio regi√≥n 0: {mean_pred_0:.2f}")
print(f"Promedio regi√≥n 1: {mean_pred_1:.2f}")
print(f"Promedio regi√≥n 2: {mean_pred_2:.2f}")
```

# 6. Calcular ganancias esperadas con los 200 mejores pozos de cada regi√≥n

```python
def calcular_ganancia(predicciones, reales):
    # Crear DataFrame con predicciones y valores reales
    df = pd.DataFrame({'pred': predicciones, 'real': reales})
    
    # Seleccionar los 200 con mayor predicci√≥n
    top_200 = df.sort_values(by='pred', ascending=False).head(200)
    
    # Calcular ganancia
    ingreso_total = top_200['real'].sum() * INGRESO_POR_UNIDAD
    ganancia = ingreso_total - BUDGET
    
    return ganancia

# Calcular ganancia esperada para cada regi√≥n
ganancia_0 = calcular_ganancia(predictions_0, y_valid_0)
ganancia_1 = calcular_ganancia(predictions_1, y_valid_1)
ganancia_2 = calcular_ganancia(predictions_2, y_valid_2)

print(f"Ganancia regi√≥n 0: ${ganancia_0:,.0f}")
print(f"Ganancia regi√≥n 1: ${ganancia_1:,.0f}")
print(f"Ganancia regi√≥n 2: ${ganancia_2:,.0f}")
```

<b>Comentario del revisor</b> <a class="tocSkip"></a><br>  
<b>√âxito</b> - Excelente implementaci√≥n del c√°lculo de ganancias. La l√≥gica para seleccionar los mejores pozos est√° bien aplicada y los resultados son claros y relevantes para la toma de decisiones.  
</div>


# 7. Evaluar riesgos y ganancias con bootstrapping
Usamos bootstrapping para analizar la distribuci√≥n de ganancias y el riesgo de p√©rdida por regi√≥n.

```python
def bootstrapping(predicciones, reales, n_simulaciones=1000):
    ganancias = []

    datos = pd.DataFrame({'pred': predicciones, 'real': reales})

    for _ in range(n_simulaciones):
        muestra = datos.sample(n=500, replace=True)
        top_200 = muestra.sort_values(by='pred', ascending=False).head(200)
        ganancia = top_200['real'].sum() * INGRESO_POR_UNIDAD - BUDGET
        ganancias.append(ganancia)

    ganancias = pd.Series(ganancias)
    media = ganancias.mean()
    intervalo = ganancias.quantile([0.025, 0.975])
    riesgo = (ganancias < 0).mean()

    return media, intervalo, riesgo
```

```python
media_0, intervalo_0, riesgo_0 = bootstrapping(predictions_0, y_valid_0)
media_1, intervalo_1, riesgo_1 = bootstrapping(predictions_1, y_valid_1)
media_2, intervalo_2, riesgo_2 = bootstrapping(predictions_2, y_valid_2)

print(f"Regi√≥n 0 - Ganancia media: ${media_0:,.0f}, Intervalo 95%: {intervalo_0.values}, Riesgo: {riesgo_0:.2%}")
print(f"Regi√≥n 1 - Ganancia media: ${media_1:,.0f}, Intervalo 95%: {intervalo_1.values}, Riesgo: {riesgo_1:.2%}")
print(f"Regi√≥n 2 - Ganancia media: ${media_2:,.0f}, Intervalo 95%: {intervalo_2.values}, Riesgo: {riesgo_2:.2%}")
```

- La regi√≥n 1 presenta el **menor riesgo de p√©rdida (1.4%)**, inferior al umbral del 2.5% exigido.
- Su ganancia media ajustada al riesgo es de aproximadamente **4.33 millones de d√≥lares**.
